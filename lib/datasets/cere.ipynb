```python
import os
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# 得到包含所有SME的id，放在Series中
path = "D:/programs/grin/datasets/irish_cere/"
id_with_code = pd.read_csv(os.path.join(path, 'IDwithCode.csv'))
id_sme = id_with_code[id_with_code['Code']==2]['ID'].tolist()
```

```python
# 创建一个空的数据字典
data_dict = {}

# 循环处理六个文件
for i in range(1, 7):
    file_name = os.path.join(path, f'File{i}.txt')
    with open(file_name, 'r') as file:
        for line in file:
            meter_id, five_digit_code, consumption = map(float, line.split())
            meter_id = int(meter_id)
            if meter_id in id_sme:
                day_code = int(five_digit_code // 100)  # 提取前三位作为Day code
                time_code = int(five_digit_code % 100) - 1  # 提取后两位并减去1，以匹配时间段
                date = datetime(2009, 1, 1) + timedelta(days=day_code - 1, minutes=time_code * 30)
                time_str = date.strftime("%Y-%m-%d %H:%M:%S")
                if time_str not in data_dict:
                    data_dict[time_str] = {}
                data_dict[time_str][str(meter_id)] = consumption

# 创建DataFrame
df = pd.DataFrame(data_dict).T
df.sort_index(axis=1, inplace=True)
df
```

```python
df.sort_index(axis=0, inplace=True)
df
```

```python
# 将DataFrame写入HDF5文件
file_name = os.path.join(path, 'cere_sme.h5')
df.to_hdf(file_name, key='data', mode='w')
# 将DataFrame写入HDF5文件并进行压缩编码
file_name = os.path.join(path, 'cere_sme_compressed.h5')
df.to_hdf(file_name, key='data', mode='w', format='fixed', complevel=3, complib='blosc')
```

```python
# 统计DataFrame中所有元素的个数
total_elements = df.size

# 统计DataFrame中值为NaN的个数
nan_count = df.isna().sum().sum()
print("总元素个数:", total_elements)
print("值为NaN的个数:", nan_count)
print("值为NaN占比:", nan_count/total_elements)
```
