```python
import os
import pandas as pd
import numpy as np

# 得到包含所有SME的id，放在Series中
path = "D:/programs/grin/datasets/lowa/"
data_feederA = pd.read_csv(os.path.join(path, 'feederA.csv'))
data_feederB = pd.read_csv(os.path.join(path, 'feederB.csv'))
data_feederC = pd.read_csv(os.path.join(path, 'feederC.csv'))
```

```python
import re
```

```python
df_feederA = pd.DataFrame(data_feederA)
df_feederA['Time'] = pd.to_datetime(df_feederA['Time'], format='%m/%d/%y %I:%M %p')
df_feederA.set_index('Time', inplace=True)
df_feederB = pd.DataFrame(data_feederB)
df_feederB['Time'] = pd.to_datetime(df_feederB['Time'], format='%m/%d/%y %I:%M %p')
df_feederB.set_index('Time', inplace=True)
df_feederC = pd.DataFrame(data_feederC)
df_feederC['Time'] = pd.to_datetime(df_feederC['Time'], format='%m/%d/%y %I:%M %p')
df_feederC.set_index('Time', inplace=True)
df = pd.concat([df_feederA, df_feederB, df_feederC], axis=1)
df
```

```python
def extract_numbers_from_column_name(column_name):
    numbers = re.findall(r'\d+', column_name)  # 使用正则表达式提取数字部分
    if numbers:
        return ''.join(numbers)  # 将提取的数字部分连接起来
    else:
        return column_name  # 如果没有数字部分，则保持原始列名
```

```python
df.columns = [extract_numbers_from_column_name(col) for col in df.columns]
```

```python
df
```

```python
file_name = os.path.join(path, 'lowa.h5')
df.to_hdf(file_name, key='data', mode='w')
```
